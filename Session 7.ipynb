## ***Why do we use for loop in python?***

***A for loop in Python is used to iterate over a sequence (like a list, tuple, string, or range) and execute a block of code repeatedly for each item in the sequence. It helps automate repetitive tasks, making the code more concise and efficient***

```python
# Tuple
my_tuple = (1, 2, 3, 4, 5)
my_tuple
```

```python
print(my_tuple[0])
print(my_tuple[1])
print(my_tuple[2])
print(my_tuple[3])
print(my_tuple[4])
```

```python
for i in my_tuple:
    print(i)
```

```python
# List
my_list = ['apple', 'banana', 'cherry']
my_list
```

```python
for x in my_list:
    print(x)
```

```python
# String
my_string = "hello"
my_string
```

```python
for a in my_string:
    print(a)
```

```python
for city in cities:
    print(city)
```

## ***Nested for loop***

```python
visited_cities = ['islamabad', 'multan','faisalabad','quetta','peshawar']
visited_cities
```

```python
cities
```

```python
for city in cities:
    for visited_city in visited_cities:
        print(city, visited_city)
```

```python
week_days = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday']
time_slots = ['9AM-10AM', '10AM-11AM', '11AM-12PM']
```

```python
for week_day in week_days:
    for time_slot in time_slots:
        print(week_day, time_slot)
```

```python
for week_day in week_days: #outer loop
    print(week_day)
    for time_slot in time_slots: #inner loop
        print(time_slot)
    print()
```

## ***while loop***

```python
num = 0

while num < 5:
    print(num)
```

```python
num = 0

while num < 5:
    print(num)
    num += 1
```

```python
num = 0

while num <= 5:
    print(num)
    num += 1
```

```python
cities_to_visit = []
user_input = ''

while user_input != 'quit':
    city = input('enter a city')
    if city == 'quit':
        user_input = 'quit'
    else:
        cities_to_visit.append(city)
print(cities_to_visit)
```

## ***Nested While loop***

```python
cities = ('karachi', 'hyderabad', 'islamabad', 'quetta', 'lahore')
visited_cities = ('islamabad', 'multan', 'faisalabad', 'quetta', 'peshawar')

i = 0

while i < len(cities):
    j = 0
    while j < len(visited_cities):
        print(cities[i], visited_cities[j])
        j += 1
    i += 1
```
