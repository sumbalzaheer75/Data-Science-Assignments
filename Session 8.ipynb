#### Write a function that accepts marks of any number of students and prints the total marks.

```python
def total_marks(*marks):
    print("Total Marks:", sum(marks))

total_marks(80, 75, 90, 85)
```

#### Write a function that accepts prices of any number of products and prints the highest price.

```python
def highest_price(*prices):
    print("Highest Price:", max(prices))

highest_price(250, 400, 150, 600, 350)
```

#### Write a function that accepts any number of temperatures and prints the average temperature.

```python
def average_temp(*temps):
    avg = sum(temps) / len(temps)
    print("Average Temperature:", avg)

average_temp(30, 32, 28, 35, 31)
```

#### Write a function that accepts any number of shopping items and prints each item.

```python
def shopping_list(*items):
    for item in items:
        print(item)

shopping_list("Milk", "Bread", "Eggs", "Butter")
```

#### Write a function that accepts employee details using keyword arguments and prints each detail.

```python
def employee_details(**details):
    for key, value in details.items():
        print(key, ":", value)

employee_details(name="Ali", age=28, department="IT")
```

#### Write a function that accepts product information using keyword arguments and prints all the details.

```python
def product_info(**product):
    for key, value in product.items():
        print(key, ":", value)

product_info(name="Laptop", brand="Dell", price=120000, stock=15)
```

A company wants to calculate an employee's net salary. Write a Python function calculate_salary() that takes:

basic_salary as a required argument,
any number of allowances (e.g., house allowance, transport allowance, medical allowance) using *args,
any number of deductions (e.g., tax, insurance, loan) using **kwargs.

The function should:

Calculate the total allowances.
Calculate the total deductions.
Calculate the gross salary.
Calculate the net salary.
Display the basic salary, total allowances, total deductions, and net salary.

```python
def calculate_salary(basic_salary, *allowances, **deductions):
    
    total_allowance = sum(allowances)
    total_deduction = sum(deductions.values())
    
    gross_salary = basic_salary + total_allowance
    net_salary = gross_salary - total_deduction
    
    print("Basic Salary:", basic_salary)
    print("Total Allowances:", total_allowance)
    print("Total Deductions:", total_deduction)
    print("Net Salary:", net_salary)
    
    # return net_salary
```

```python
calculate_salary(
    50000,
    5000, 3000, 2000,
    tax=4000,
    insurance=1500,
    loan=1000
)
```

#### Write a Python program using a nested function. The outer function should accept student marks, and the inner function should calculate the grade.

```python
def student_result(marks):

    def calculate_grade():
        if marks >= 80:
            return "A"
        elif marks >= 60:
            return "B"
        else:
            return "C"

    print("Marks:", marks)
    print("Grade:", calculate_grade())

student_result(75)
```

#### Write a Python program using a nested function. The outer function should accept the basic salary, and the inner function should calculate a 15% bonus. Print basic salary, bonus, and total(basic + bonus)

```python
def salary(basic):

    def bonus():
        return basic * 0.15

    print("Basic Salary:", basic)
    print("Bonus:", bonus())
    print("Total Salary:", basic + bonus())

salary(50000)
```
